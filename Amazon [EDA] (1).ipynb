# 3. Data Wrangling

## Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Drop duplicate and less columns for visualization
columns_to_keep = [
    'id', 'title', 'type', 'description', 'release_year', 'age_certification',
    'runtime', 'genres', 'production_countries', 'seasons',
    'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score'
]
df_reduced = data[columns_to_keep].drop_duplicates(subset='id')

# Handle missing values
df_reduced['age_certification'].fillna('Unknown', inplace=True)
df_reduced['description'].fillna('No description available.', inplace=True)
df_reduced['seasons'].fillna(0, inplace=True)
df_reduced['imdb_score'].fillna(df_reduced['imdb_score'].median(), inplace=True)
df_reduced['imdb_votes'].fillna(0, inplace=True)
df_reduced['tmdb_popularity'].fillna(0, inplace=True)
df_reduced['tmdb_score'].fillna(df_reduced['tmdb_score'].median(), inplace=True)

# Convert relevant columns to appropriate types
df_reduced['release_year'] = df_reduced['release_year'].astype('Int64')
df_reduced['seasons'] = df_reduced['seasons'].astype('int')
df_reduced['genres'] = df_reduced['genres'].str.strip("[]").str.replace("'", "").str.split(", ")

# Display cleaned dataset
df_reduced.info(), df_reduced.head()
```

# chart - 4 [Runtime Distribution]

```python
sns.histplot(data['runtime'], bins=50, kde=True)
plt.title("Distribution of Runtime")
plt.xlabel("Runtime (minutes)")
plt.show()
```

# chart - 5 [IMDB score distribution]

```python
sns.histplot(data['imdb_score'], bins=20, kde=True)
plt.title("IMDb Score Distribution")
plt.xlabel("Score")
plt.show()
```

# chart - 6 [TMDB vs IMDB Score]

```python
sns.scatterplot(x='tmdb_score', y='imdb_score', hue='type', data=data)
plt.title("TMDB vs IMDb Score")
plt.show()
```

# chart - 7 [Count by Title by Country]

```python
data['production_countries'] = data['production_countries'].astype(str).str.strip("[]").str.replace("'", "")
country_counts = data['production_countries'].value_counts().head(10)
country_counts.plot(kind='bar')
plt.title("Top 10 Production Countries")
plt.ylabel("Number of Titles")
plt.show()

```
